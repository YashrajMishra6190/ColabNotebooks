# Regex: Beginners Guide to Regular Expression
---
A `RegEx` is a powerful tool for matching text, based on a predefined pattern. It can detect the presence or absence of a text by matching it with a particular pattern, and also can split a pattern into one or more sub-patterns. The Python standard library provides a re module for regular expressions. Its primary function is to offer a search, where it takes a regular expression and a string. Here, it either returns the first match or else none.

##1. Search

```python
# re helps to make different kinds of patterns: we have text and had to extract only the emails so we use re(regular expression)
import re
```

```python
txt = 'Hello, How are you?'
# check if you is there or not
'you' in txt
```

```python
# Now for the same string
srch = re.search('you', txt)
srch
#span = indexing(start, end-1)
```

```python
srch.span()
```

```python
#Starting index
print(srch.start())
#Ending index
print(srch.end())
```

##4. Creating Regular Expressions

```python
# \d - digits
txt2 = 'My telephone number is 883-9886-789'

# pattern = '\d\d\d\d\d\d\d\d\d\d'
pattern = '\d{3}-\d{4}-\d{3}'
re.search(pattern, txt2)
```

```python
# \w - alphanumeric
pattern2 = '\w\w'
re.search(pattern2, txt2)
```

```python
# To print all the values .group()
re.search(pattern2, txt2).group()
```

```python
re.search(pattern, txt2).group()
```

# Pattern Matching with Alphanumeric

```python
import re
```

##1. Finding pattern with specific length ending

```python
re.findall('at', 'The rat sat on the mat and attached by a cat')
```

```python
# if you want to see the word that contains specific characters so add '.'
re.findall('.at', 'The rat sat on the mat and attached by a cat')
```

```python
# for two characters add '..'
re.findall('..at', 'The rat sat on the mat and attached by a cat')
```

##2. Start and End Pattern Matching

```python
re.findall('\d', '4 is divisible by 2 not by 3')
```

```python
# \d$ here $ returns number present at the end if not returns []
re.findall('\d$', '4 is divisible by 2 not by 3')
```

```python
# ^\d here ^ returns number present at the start if not returns []
re.findall('^\d', '4 is divisible by 2 not by 3')
```

##3. Finding Capital Letters

```python
# Here I mention the range of capital letters from [A-Z]
re.findall('[A-Z]', 'Welcome To Amity')
```

##4. Finding Small Letters

```python
# Here I mention the range of small letters from [a-z]
re.findall('[a-z]', 'Welcome To Amity')
```

##5. Finding Uppercase and Lowercase

```python
# mention the range after ranges
re.findall('[A-Za-z]', 'Welcome To Amity')
```

##6. Finding Uppercase, Lowecase and Numbers

```python
re.findall('[A-Za-z0-9]', 'Welcome To Amity 1')
```

# Text Preprocessing with RE

```python
import re
```

##1. Removing the Special Characters

```python
txt = 'Eshant is happy! But why is he not playing?'
```

```python
#excluded
re.findall('[^!]+', txt)
```

```python
# to Exclude all the special characters  '[^...]' after ^ mention all the special characters that you want to exclude
# also after []+ here + dentoes to print all the characters at a single line
txt = 'Eshant is happ&y! But wh*y is h$e no*t playing?'
re.findall('[^$&!*?]+', txt)
```

```python
# to make more readable format use concatenation .join()
''.join(re.findall('[^$&!*?]+', txt))
```

##2. Exclusion

Exclusion says that
- print all the characters except digits{numbers}; we will write \D instead of \d
- print all the special characters accept alpha numeric characters; we will write \W instead of \w

```python
txt = "I'm 20"
"".join(re.findall('\D', txt))
```

```python
# Similarly, for the alphanumeric exclusion
"".join(re.findall('\W', txt))
```

# Email Pattern Matching with re

```python
import re
```

```python
mail = 'yashrajmishra619@gmail.com'
p = '[A-Za-z0-9]+@[\w]+.[\w]+'
re.search(p,mail)
```

```python
re.search(p, 'theyashorg@gmail.com').group()
```

##1. Email ID with specific Domain

```python
student_mail = 'yashraj@s.amity.edu'
pattern = '[A-Za-z0-9]+@(s).(amity).(edu)'
re.search(pattern, student_mail)
```

##2. Email ID with Multiple selected Domain

```python
student_mail = 'yashraj@s.amity.org'
pattern = '[A-Za-z0-9]+@[A-Za-z0-9].(amity).(edu|org|gov)'
re.search(pattern, student_mail)
```
