# ASCII (American Standard Code for Information Interchange)


---


It is a standard character encoding used in telecommunication. The ASCII pronounced ‘ask-ee’ , is strictly a seven bit code based on English alphabet. ASCII codes are used to represent alphanumeric data . The code was first published as a standard in 1967. it was subsequently updated and published as  ANSI X3.4-1968, then as ANSI X3.4-1977 and finally as ANSI X3.4-1986. Since it is a seven bit code , it can at the most represent 128 characters. it currently defines 95 printable characters including 26 upper case letters (A to Z) , 26 lower case letters , 10 numerals (0 to 9) and 33 special characters including mathematical symbols, punctuation marks and space character. They represent text in, telecommunications equipment, and devices. These include numbers, upper and lowercase English letters, functions, punctuation symbols, and some other symbols. In total, there are 256 ASCII characters, and can be broadly divided into three categories:

ASCII control characters (0-31 and 127)
ASCII printable characters (32-126) (most commonly referred)
Extended ASCII characters (128-255)



# ord() Function


---
Ord function is used to convert a single uunicode character into its integer represesentation, `only takes single characters at a time.`


```python
"""
Natural Language are nothing but strings. And this strings are represented as ASCII list of characters in words. So like we had a name = "Yashraj Mishra" so computer will read as the ASCII code version.
"""

#For example;
name = "yashraj mishra"
for char in name:
  print(char, ord(char))
```

```python
#now lets see for the capital letters
cap_name = "YASHRAJ MISHRA"
for ch in cap_name:
  print(ch, ord(ch))
```

```python
# Now, How they are represented its because of the ord() function which represents all the characters in ASCII format
#For small letter
corpus_alpha = "abcdefghijklmnopqrstuvwxyz"
for char in corpus_alpha:
  print(char, ord(char))
```

```python
#For Capital Letters
corpus_alpha_cap = corpus_alpha.upper()
for char in corpus_alpha_cap:
  print(char, ord(char))
```

```python
#For numberic value
numbers = "0123456789"
for char in numbers:
  print(char, ord(char))
```

```python
# for the representation: we can use comprehension
[ord(ch) for ch in numbers]
```

```python
#For Special Characters

sp_ch = "~!@#$%^&*()_+=-[]{}\|;':<>,.?/`"
for char in sp_ch:
  print(char, ord(char))
```

# DIY Functions: capitalize()


---

Python String capitalize() method returns a copy of the original string and converts the first character of the string to a capital (uppercase) letter, while making all other characters in the string lowercase letters.

```python
text = "the quick brown fox jumps over the lazy dog."
text
```

Using the `capitalize() function` only the first letter/character  of the sentence will be upper case and rest of the letter will be lower case.

```python
text.capitalize()
```

```python
"""

Let's see the naive approach behind the capitalize().
First, the divide the characters of the text take 1 character and then take whole characters
Secondly, provide the upper() function in the first character and then print the text as it is.
So this is the naive approach followed by the capitalize() function.

"""
#First character
print(text[0].upper())

# Second Characters: rest characters
print(text[1:])
```

```python
# logic behind the capitalize function
text_capitalize = text[0].upper() + text[1:]

text_capitalize
```

```python
# Check for the uppercase characters

for char in txt:
  if (ord(char)>= ord('A') and ord(char)<= ord('Z')):
    print(char, ord(char))
  elif(ord(char)>= ord('a') and ord(char)<= ord('z')):
    print(char, ord(char))
  else:
    print(char)
```

```python
# Now, the difference between ASCII code of lowercase and uppercase
ord('a')-ord('A')
```

```python
ord('c') - ord('C')
```

```python
# So Now converting the lowercase letters in the ascii code by subtracting 32 from the ASCII code
for char in txt:
  if (ord(char) >= ord('a') and ord(char) <= ord('z')):
    print("Lowercase -> ", char, "-> Uppercase -> ", chr(ord(char)-32))
  else:
    print(char)
```

```python
#Optimized code
def upper(txt):
  string = ''
  for char in txt:
    if (ord(char) >= ord('a') and ord(char) <= ord('z')):
      string += chr(ord(char)-32)
    else:
      string += char
  print(string)
```

```python
upper(txt)
```

```python
#Alternate code for the same,
txt.upper()
```

# Lowercase Function

```python
def lower(txt):
  string = ''
  for char in txt:
    if (ord(char) >= ord('A') and ord(char) <= ord('Z')):
      string += chr(ord(char)+32)
    else:
      string += char
  print(string)
```

##1. isalpha()

```python
txt = 'Yashraj Mishra'
txt
```

```python
def isalpha(txt):
  count = 0
  for char in txt:
    if((ord(char) >= ord('A') and ord(char)<= ord('Z')) or (ord(char) >= ord('a') and ord(char)<= ord('z'))):
      count+=1

  print(count)
  if (len(txt) == count):
    return True
  else:
    return False
```

##2. isdigit()

```python
def isdigit(txt):
  count = 0
  for char in txt:
    if(ord(char) >= ord('0') and ord(char) <= ord('9')):
      count += 1
  if (len(txt) == count):
    return True
  else:
    return False
```

```python
def title_(txt):
  str = ''
  for word in txt.split(' '):
    str += upper(word[0]) + lower(word[1:])
  return str
```

```python
# Optimized Way
def title(text):
  output = []
  for word in text.split(' '):
    output.append(upper(word[0]) + lower(word[1:]))

  return ' '.join(output)
```
